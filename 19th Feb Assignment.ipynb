###  What is Flask Framework?

Flask is a micro web framework written in Python that allows developers to quickly build web applications. It is lightweight, easy to learn, and highly customizable. Flask provides a variety of tools and libraries that enable developers to create web applications with minimal code and configuration. It is often used for building small to medium-sized web applications such as simple websites, blogs, and APIs. Flask is based on the Werkzeug WSGI toolkit and the Jinja2 template engine, and it supports HTTP requests and responses, URL routing, cookie handling, and more

### What are the advantages of Flask Framework?

1. Easy to learn and use: Flask has a simple and intuitive interface that makes it easy for developers to learn and use. It also provides extensive documentation and community support.

2. Lightweight: Flask is a micro-framework that has a small footprint and minimal dependencies, making it easy to deploy and run on various platforms.

3. Flexible: Flask allows developers to customize and extend its functionality according to their requirements. It also supports a wide range of extensions and libraries for various tasks.

4. Modular design: Flask has a modular design that allows developers to use only the components they need, making it easy to build and maintain applications.

5. Ideal for small to medium-sized applications: Flask is best suited for small to medium-sized applications that do not require the complexity and features of a full-stack framework like Django.

6. Built-in development server: Flask comes with a built-in development server, making it easy to develop and test applications locally.

7. Compatible with various databases: Flask supports a wide range of databases, including SQLite, MySQL, PostgreSQL, and MongoDB.

### What is App routing in Flask?

App routing in Flask refers to the process of mapping the URL of a web page to a specific function or view that will handle the request and generate a response for that page. In other words, it is a way to define which function should be executed for a given URL or route. This is done using the @app.route() decorator in Flask.

### Why do we use app routes?

We use app routes in Flask to define the URLs that will trigger the execution of a specific function or view. These routes help the Flask application to respond to the client's requests by defining the endpoints or URLs for the application. The routes are used to match the requested URL pattern and execute the corresponding function or view, which generates the HTTP response to be returned to the client.

In summary, app routes are used to:

1. Define the URLs or endpoints for the Flask application.
2. Trigger the execution of a specific function or view when a specific URL is requested.
3. Map the requested URL pattern to a corresponding function or view.
4. Help the Flask application to respond to the client's requests.

### What function is used in Flask for URL Building? 

In Flask, the url_for() function is used for URL building. It generates a URL to the given endpoint with the method provided. It takes the name of the endpoint as its first argument and any number of keyword arguments, each corresponding to a variable part of the URL rule for the endpoint.

### Write a Python code to demonstrate the working of the url_for() function.

```python
from flask import Flask, redirect, url_for

appFlask = Flask(__name__)

@appFlask.route('/accountType/<Type>')
def accountType(Type):
    return 'This is a %s account' % Type

@appFlask.route('/user/<name>')
def userType(name):
    if name =='premium':
        return redirect(url_for('accountType',Type = name))
    else:
        return redirect(url_for('accountType',Type = name))
    
if __name__ == '__main__':
    appFlask.run(debug = True)
```
