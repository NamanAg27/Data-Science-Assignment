### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a data visualization library for Python programming language. It allows developers to create various types of static, animated, and interactive visualizations in Python. Matplotlib is used to create charts, graphs, and other visual representations of data. It is a powerful tool for visualizing large datasets and presenting them in a way that is easy to understand.

Some of the popular plots that can be created using the Pyplot module of Matplotlib include:

1. Line Plot: A line plot displays data points connected by lines. It is commonly used to show trends over time.

2. Scatter Plot: A scatter plot displays data points as individual dots. It is commonly used to show the relationship between two variables.

3. Bar Chart: A bar chart displays data as rectangular bars. It is commonly used to compare data across different categories.

4. Histogram: A histogram displays the distribution of a dataset. It is commonly used to show the frequency of values in a dataset.

5. Pie Chart: A pie chart displays data as a circular chart divided into slices. It is commonly used to show the proportion of different categories in a dataset.

### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

### Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a type of plot in Matplotlib that displays the values of two variables as a collection of points on a two-dimensional plane. Each point in the scatter plot represents the values of the two variables for a single observation.

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
plt.xlabel("This is the X-Axis")
plt.ylabel("This is the Y-Axis")
plt.title("Scatter Plot Assignment")
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.figure(figsize=(7,7))
plt.subplot(2,2,1)
plt.plot(x1,y1,c="r")

plt.subplot(2,2,2)
plt.plot(x2,y2,c="g")

plt.subplot(2,2,3)
plt.plot(x3,y3,c="b")

plt.subplot(2,2,4)
plt.plot(x4,y4,c="y")

plt.show()
```

The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to divide a single figure into multiple subplots, each with its own set of axes.

The subplot() function takes three integer arguments: the number of rows, the number of columns, and the index of the subplot. The index starts at 1 and increases left to right, top to bottom.

### Q4. What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

Bar plots are used to compare values across different categories or groups. They are particularly useful for showing data that can be divided into discrete categories or groups, such as the number of sales for different products or the average score of different teams in a competition. Bar plots can also be used to display continuous data that has been divided into groups or ranges, such as the distribution of ages in a population.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.bar(company,profit,color="y")

plt.subplot(1,2,2)
plt.barh(company,profit,color="g")
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot is a type of graphical display that provides a summary of a dataset's distribution. It uses a rectangular box and "whiskers" to show the range, variability, and outliers of the data.

The box in a box plot represents the middle 50% of the data, with a line inside indicating the median (the value separating the top and bottom 50%). The "whiskers" extend to the minimum and maximum values within a certain range of the box, and any data points beyond the whiskers are shown as individual dots or asterisks, representing outliers.

Box plots are especially useful when the data is not normally distributed, as they provide a clearer picture of the spread and skewness of the data compared to other types of plots, such as histograms or density plots. They can also help to identify any unusual or extreme values in the dataset that may need further investigation.

Overall, box plots are a powerful tool for exploratory data analysis and visualization, providing a quick and easy way to summarize and compare the distribution of a variable across different groups or categories.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.boxplot(box1)

plt.subplot(1,2,2)
plt.boxplot(box2)
```
